## Análise e Gerenciamento do Inventário de Anúncios

**Objetivo:** Avaliar e organizar o inventário de anúncios existentes para garantir que eles estão sendo exibidos corretamente e são relevantes.

Iremos fazer isso mapeando os anúncios atuais, análisando a Performance e por fim recomendando otimizações.

````python
import pandas as pd

inventario = pd.read_csv("C:/Users/welly/Downloads/Projeto_Ifood/inventario_anuncios.csv")

print(inventario.head())

ctr_medio = inventario.groupby('Categoria')['CTR'].mean()
print(ctr_medio)

anuncios_ineficientes = inventario[inventario['CTR'] < 0.05]
print(anuncios_ineficientes)
````

### Explicação do Teste A/B Realizado

O teste A/B foi conduzido para comparar o desempenho de dois tipos de anúncios: **Texto** e **Imagem**, avaliando os indicadores **CTR** (Click-Through Rate) e **CPA** (Custo por Aquisição). O teste foi realizado em dois períodos distintos.

#### 1. **Estrutura do Teste A/B:**
   O teste A/B foi dividido em duas fases:
   - **Fase 1 (2025-01-01)**: Anúncios com formato **Texto** e **Imagem**.
   - **Fase 2 (2025-01-10)**: Repetição do teste com os mesmos tipos de anúncios.

#### 2. **Descrição dos Dados:**

   - **Imagem**: 0.20

   Isso significa que os anúncios do tipo **Imagem** tiveram um desempenho ligeiramente melhor em termos de cliques.

#### 4. **Média de CPA por Tipo de Anúncio:**
   A média de **CPA** (Custo por Aquisição) por tipo de anúncio foi:

   - **Imagem**: 4.55
   - **Texto**: 4.75

   O anúncio de **Imagem** teve um custo mais baixo para adquirir um novo cliente em comparação com o anúncio de **Texto**.

#### 5. **Resultado Final do Teste:**
   - **CTR do Anúncio de Texto**: 0.175
   - **CTR do Anúncio de Imagem**: 0.20

   Os resultados indicam que, embora o **CPC** tenha sido um pouco mais alto para os anúncios de **Imagem**, esses anúncios apresentaram um **CTR** superior, o que pode ser mais vantajoso em termos de engajamento.

Este gráfico de barras representa a distribuição de feedbacks categorizados por nível de prioridade.  

### **Eixos**  
- **Eixo X (Prioridade):**  
  - Classifica os feedbacks em três categorias: **Alta, Média e Baixa**.  

- **Eixo Y (Quantidade de Feedbacks):**  
  - Indica o número total de feedbacks em cada categoria.  

### **Interpretação da Distribuição**  
- **Prioridade Alta:**  
  - A maior quantidade de feedbacks está nesta categoria, com **14 registros**.  
  - Indica que há muitas questões críticas que precisam de atenção imediata.  

- **Prioridade Média:**  
  - Representa **9 feedbacks**, indicando uma quantidade intermediária de itens que exigem atenção, mas sem urgência máxima.  

- **Prioridade Baixa:**  
  - Apenas **3 feedbacks**, sugerindo que há poucos pontos considerados menos urgentes.  

## **Principais Insights**  
1. **Foco em Questões Críticas**  
   - O alto volume de feedbacks na categoria **Alta** sugere que muitas demandas requerem ação prioritária.  

2. **Balanceamento de Tarefas**  
   - A categoria **Média** também tem um número significativo de feedbacks, o que pode indicar a necessidade de planejamento para evitar acúmulo de problemas urgentes no futuro.  

3. **Poucas Questões de Baixa Prioridade**  
   - O número reduzido de feedbacks na categoria **Baixa** pode indicar que a maioria das questões levantadas tem impacto relevante e exigem ação rápida.  

## **Conclusão**  
A predominância de feedbacks de **Alta e Média prioridade** sugere que há muitas demandas urgentes a serem tratadas. Um plano de ação estruturado deve ser adotado para resolver os problemas mais críticos primeiro e evitar que os de prioridade média evoluam para urgentes.  


# Análise do Gráfico: CTR vs CPA dos Anúncios  

## **Descrição do Gráfico**  
Este gráfico de dispersão representa a relação entre **CTR (Click-Through Rate)** e **CPA (Custo por Aquisição)** dos anúncios analisados.  

### **Eixos**  
- **Eixo X (CTR - Click-Through Rate):**  
  - Mede a taxa de cliques dos anúncios, indicando sua capacidade de atrair a atenção do público.  
  - Os valores variam aproximadamente de 0 até 0,9, com a maioria dos pontos concentrados em valores baixos (próximos de 0,05).  

- **Eixo Y (CPA - Custo por Aquisição):**  
  - Mede o custo médio para converter um usuário após o clique no anúncio.  
  - Os valores variam entre aproximadamente 2,5 e 7,5.  

### **Interpretação da Distribuição**  
- **Pontos Azuis:** Representam **todos os anúncios**.  
- **Pontos Vermelhos:** Destacam **anúncios ineficazes (CTR < 0,05)**.  
- **Maior Concentração:**  
  - Observa-se uma alta densidade de pontos na parte inferior esquerda, indicando que muitos anúncios têm CTR baixo e CPA elevado.  
- **Correlação:**  
  - Há uma tendência de que anúncios com CTR mais baixo apresentem CPA mais alto, sugerindo ineficiência na conversão.  

## **Principais Insights**  
1. **Otimização Necessária**  
   - Muitos anúncios apresentam CTR baixo e CPA alto, o que pode indicar problemas no design, segmentação ou mensagem.  

2. **Atenção aos Anúncios Ineficazes**  
   - Os anúncios destacados em vermelho possuem um desempenho muito fraco e devem ser ajustados.  

3. **Melhorias Potenciais**  
   - Realizar testes A/B para testar diferentes títulos, imagens e segmentações.  
   - Revisar estratégias para aumentar a taxa de cliques e reduzir o custo por aquisição.  

## **Conclusão**  
Essa análise sugere que melhorar o **CTR** dos anúncios pode levar a uma **redução do CPA**, tornando as campanhas mais eficientes. Estratégias baseadas em dados são fundamentais para otimizar investimentos em publicidade digital.

### Explicação do Conjunto de Dados de Anúncios e Testes A/B

Este conjunto de dados contém informações detalhadas sobre os anúncios em uma campanha publicitária, além de dados sobre testes A/B realizados. Ele inclui dados sobre **ID do Anúncio**, **Tipo de Anúncio**, **Categoria**, **Status**, **CTR**, **CPA**, **Datas de Início e Fim**, **Stakeholder**, **Requisitos**, **Prioridade**, e os resultados de testes A/B com **CTR**, **CPC**, **CPA**, e **Data do Teste**.

#### 1. **Descrição das Colunas:**
   - **ID do Anúncio**: Identificador único de cada anúncio.
   - **Tipo de Anúncio**: Tipo de mídia utilizada, como Banner ou Carrossel.
   - **Categoria**: A categoria do anúncio, como Promoções ou Restaurantes.
   - **Status**: O status do anúncio, se está **Ativo** ou **Inativo**.
   - **CTR**: Taxa de cliques do anúncio, que mede o engajamento.
   - **CPA**: Custo por ação, que indica o custo por conversão do anúncio.
   - **Data de Início**: Quando o anúncio começou a ser veiculado.
   - **Data de Fim**: Quando o anúncio terminou a veiculação.
   - **Stakeholder**: O responsável pelo requisito, como Marketing, Produto ou Vendas.
   - **Requisito**: A descrição do requisito relacionado ao anúncio.
   - **Prioridade**: A prioridade do requisito (Alta, Média, Baixa).
   - **Status do Requisito**: O progresso do requisito (Em andamento, Concluído).
   - **Data de Reunião**: Data da reunião relacionada ao requisito.
   - **Teste A/B**: Teste realizado para comparar diferentes tipos de anúncio.
   - **CPC**: Custo por clique durante o teste A/B.
   - **Data do Teste**: A data em que o teste A/B foi realizado.
